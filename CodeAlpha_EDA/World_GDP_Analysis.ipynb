### Top 6 Countries

```python
lt=[]
# finding top 6 countries in 2023
top_6=(df.loc[df.index,'2023']).nlargest(6)
# adding the values from 1980 to the top 6 countries
for i in top_6.index:
    lt.append((df.loc[i,'1980':]))
```

```python
top_6_gdp = pd.DataFrame(lt)
top_6_gdp.rename_axis(index='Country',columns='Year',inplace=True)
top_6_gdp
```

```python
plt.figure(figsize=(28,12))
sns.lineplot(data = top_6_gdp.T,marker='o')
plt.legend(fontsize = 'x-large')
plt.xlabel('Year', fontsize = '20')
plt.ylabel('Nominal GDP', fontsize = "18")
plt.title('GDP Growth of Top 6 Economies Currently(FY 2023-24)', fontsize = "18")
plt.show()
```

### Compare the yearly growth rate

```python
growth_rate=top_6_gdp.pct_change(axis='columns')*100
growth_rate.fillna(0,inplace=True)
growth_rate
```

```python
plt.figure(figsize=(23,10))
sns.lineplot(data = growth_rate.T[['USA',"China", "Japan"]],marker='o')
plt.ylabel('Growth Rate (%)')
plt.title('Growth Rate YoY Comparison between USA vs China vs Japan')
plt.legend(fontsize = "18")
plt.show()
```

```python
plt.figure(figsize = (23,12))
sns.lineplot(data = top_6_gdp.T[["USA","China", "Japan"]], marker = 'o')
plt.legend(fontsize = '18')
```

```python
from scipy import stats
```

```python
valid_data = df[['1980', '2023']].dropna()
corr, p_value = stats.spearmanr(valid_data['1980'], valid_data['2023'])

print(f"Spearman Correlation (1980 vs 2023): {corr:.3f}")
print(f"P-value: {p_value:.4e}")
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=valid_data, x='1980', y='2023', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xscale('log')
plt.yscale('log')
plt.title('Persistence of Wealth: 1980 GDP vs 2023 GDP (Log-Log Scale)')
plt.show()
```
